2. Nested Loop

```javascript
// array of acceleration
let acceleartion  = [10, 20, 30, 40, 50]
//array of masses
let mass = [100, 200, 300, 400, 500]
//using nested loop to solve force
let  force = []

for (var i=0; i<acceleartion.length; i++)
{for (var j =0; j<mass.length; j++)
    {
        force.push(acceleartion[i]*mass[j])
    }
}
console.log("Total Force: ", force)
```

6. Looping Over Arrays 

```javascript
dataset = ["Lion", "Tiger", "Zebra", "Leapord", "Wolf", "Trex"]
```

```javascript
for (var i =0; i <= dataset.length; i++){
    console.log(i+1, dataset[i])
}
```

```javascript
for (let i = 0; i < dataset.length; i++) {
  if (dataset[i] === "Zebra") {
    dataset[i] = "Bison";
  }
}

console.log(dataset);
```

```javascript
for (let i = dataset.length - 1; i>=0; i-- ){
    console.log(i, dataset[i])
}
```

```javascript
//variable storing random number
let targetNum = Math.floor(Math.random()*10);
let guess = Math.floor(Math.random()*10);
```

```javascript
while(true){
    guess = Math.floor(Math.random()*10);
    if(guess == targetNum){
        console.log(`Correct! Guessed: ${guess} & target was: ${targetNum}`);
        break
    }
    else{
        console.log(`Guessed ${guess} Incorrect `)
    }

}
```
