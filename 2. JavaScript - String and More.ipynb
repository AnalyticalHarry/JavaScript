In computer programming, a string is a sequence of characters. It is a data type used to represent and manipulate textual data in many programming languages, including JavaScript.

In JavaScript, a string is a series of characters enclosed within single quotes (' '), double quotes (" "), or backticks ( ).

#### 4. Substring: Extracting a portion of a string using the substring() method.

The substring() method is used to extract a portion of a string based on specified indices. It is similar to the slice() method but has slightly different behavior when dealing with negative indices.

The substring() method takes two parameters: the start index and the end index (optional). It returns a new string that represents the extracted portion of the original string. The start index is inclusive, and the end index is exclusive, meaning the character at the end index is not included in the extracted substring.

#### charAt

The charAt() method is a built-in function in JavaScript that is used to retrieve the character at a specified index position within a string. It allows you to access a specific character within a string by providing the index as an argument.

string.charAt(index)


string: The string from which you want to extract a character.

index: The index position of the character you want to retrieve. The index starts at 0 for the first character, 1 for the second character, and so on.

```javascript
let str = 'Hello World';
console.log(str.charAt(0)); // Output: H
console.log(str.charAt(2)); // Output: l
console.log(str.charAt(4)); // Output: o
```

#### endsWith


The endsWith() method is a built-in function in JavaScript that is used to check whether a string ends with a specified substring. It returns a boolean value (true or false) indicating whether the string ends with the specified substring.

string.endsWith(searchString, length)


string: The string on which you want to perform the endsWith() check.

searchString: The substring you want to check if it appears at the end of the string.

length (optional): The length of the string to be considered for the endsWith() check. If not specified, the entire string is considered.

The endsWith() method returns true if the string ends with the specified substring, and false otherwise.

```javascript
console.log(str.endsWith('World')); // Output: true
console.log(str.endsWith('Hello')); // Output: false
```

#### indexOf

The indexOf() method is a built-in function in JavaScript that is used to find the index position of the first occurrence of a specified substring within a string. It returns the index of the substring if found, or -1 if the substring is not present in the string.

string.indexOf(searchValue, startIndex)

string: The string in which you want to search for the substring.

searchValue: The substring you want to find the index of.

startIndex (optional): The index position at which the search should begin. If not specified, the search starts from index 0.

The indexOf() method returns the index of the first occurrence of the substring within the string. If the substring is not found, it returns -1.

```javascript
console.log(str.indexOf('World')); // Output: 6
console.log(str.indexOf('John')); // Output: -1 (substring not found)
```

#### startsWith

The startsWith() method is a built-in function in JavaScript that is used to check whether a string starts with a specified substring. It returns a boolean value (true or false) indicating whether the string starts with the specified substring.

string.startsWith(searchString, position)


string: The string on which you want to perform the startsWith() check.

searchString: The substring you want to check if it appears at the beginning of the string.

position (optional): The index position at which the startsWith() check should start within the string. If not specified, it starts from index 0.

The startsWith() method returns true if the string starts with the specified substring, and false otherwise.

```javascript
console.log(str.startsWith('Hello')); // Output: true
console.log(str.startsWith('World')); // Output: false
```

#### split

The split() method is a built-in function in JavaScript that is used to split a string into an array of substrings based on a specified separator. It allows you to break a string into multiple parts using a delimiter and store those parts as separate elements in an array.

string.split(separator, limit)


string: The string you want to split into substrings.

separator: The delimiter used to determine where to split the string. It can be a string or a regular expression. If omitted, the entire string is treated as a single element in the resulting array.

limit (optional): An integer that specifies the maximum number of splits to be performed. The resulting array will have a maximum length of limit + 1. If not specified, the entire string is split.

The split() method returns an array of substrings obtained by splitting the original string based on the specified separator.

```javascript
let str__ = 'Hello,World,JavaScript';
let parts = str__.split(',');
console.log(parts); // Output: ["Hello", "World", "JavaScript"]
```

```javascript
typeof(parts)
```

#### replace

The replace() method is a built-in function in JavaScript that is used to replace occurrences of a specified substring or pattern within a string with a new substring or value. It allows you to search for a specific substring or pattern in a string and replace it with another value.

string.replace(searchValue, newValue)


string: The original string in which you want to perform the replacement.

searchValue: The substring or pattern you want to search for within the string. It can be a string or a regular expression.

newValue: The new substring or value that will replace the occurrences of searchValue within the string.

```javascript
let replaced = str.replace('World', 'Universe');
console.log(replaced); // Output: Hello Universe
```

#### sliced

The slice() method is a built-in function in JavaScript that is used to extract a portion of a string and return it as a new string. It allows you to specify the starting and ending positions to define the substring you want to extract.

string.slice(startIndex, endIndex)


string: The original string from which you want to extract a substring.

startIndex: The index position at which the extraction should begin. The character at this index will be included in the extracted substring. If negative, it represents an offset from the end of the string.

endIndex (optional): The index position immediately after the end of the extraction. The character at this index will not be included in the extracted substring. If omitted, the slice() method will extract the substring until the end of the string.

The slice() method returns a new string that contains the extracted substring.

```javascript
let sliced = str.slice(6, 11);
console.log(sliced); // Output: World
```

#### padStart

The padStart() method is a built-in function in JavaScript that is used to pad a string with another string or characters at the beginning until it reaches a specified length. It allows you to add padding to the left side of a string.

string.padStart(targetLength, padString)


string: The original string to which padding will be applied.

targetLength: The desired length of the resulting string after padding. If the original string is already longer than targetLength, no padding will be applied.

padString (optional): The string or characters to be used for padding. It will be repeated as necessary to reach the targetLength. If not specified, a space character will be used as the default padding string.

The padStart() method returns a new string with the specified padding applied.

```javascript
let message_ = 'Hello';
let paddedStart = message_.padStart(10, '-');
console.log(paddedStart); // Output: -----Hello

let paddedEnd = message_.padEnd(10, '-');
console.log(paddedEnd); // Output: Hello-----
```

#### includes

The includes() method is a built-in function in JavaScript that is used to determine whether a string contains a specified substring or not. It checks if the string includes the specified substring and returns a boolean value (true or false) based on the result.

string.includes(substring, startIndex)


string: The original string in which you want to search for the substring.

substring: The substring that you want to check for within the string.

startIndex (optional): The index position at which the search should start. If omitted, the search starts from the beginning of the string.

The includes() method returns true if the substring is found in the string, and false otherwise.

```javascript
console.log(str.includes('World')); // Output: true
console.log(str.includes('John')); // Output: false
```

#### repeat

The repeat() method is a built-in function in JavaScript that is used to create and return a new string by concatenating the original string with itself a specified number of times. It allows you to repeat a string a certain number of times.

string.repeat(count)


string: The original string that you want to repeat.

count: The number of times you want to repeat the string. It must be a non-negative integer.

The repeat() method returns a new string that consists of the original string repeated count times.

```javascript
let repeated = str.repeat(3);
console.log(repeated); // Output: Hello Hello Hello 
```

#### charCodeAt

The charCodeAt() method is a built-in function in JavaScript that is used to retrieve the Unicode value of a character at a specified index within a string. It returns an integer representing the Unicode value of the character.

string.charCodeAt(index)


string: The original string from which you want to retrieve the Unicode value.

index: The index position of the character within the string for which you want to get the Unicode value. The index starts from 0.

The charCodeAt() method returns an integer representing the Unicode value of the character at the specified index. If the index is out of range (i.e., less than 0 or greater than or equal to the length of the string), it returns NaN (Not a Number).

```javascript
console.log(str.charCodeAt(0)); // Output: 72 (Unicode value of 'H')
console.log(str.charCodeAt(4)); // Output: 111 (Unicode value of 'o')
```

#### match

The match() method is a built-in function in JavaScript that is used to search a string for a specified pattern and return an array containing all the matches found. It allows you to extract substrings that match a specified regular expression.

string.match(regexp)


string: The original string in which you want to search for the pattern.

regexp: A regular expression (regex) pattern or a RegExp object that defines the pattern to be matched against the string.

The match() method returns an array of matches found in the string. If no matches are found, it returns null.

```javascript
let strLine = 'Hello World Hello';
let matches = strLine.match(/Hello/g);
console.log(matches); // Output: ["Hello", "Hello"]
```

### Undefined & Null

Null are Intentional absence of any value and must be assigned.

Undefined are variables that do not have an assigned value are undefined. 

```javascript
let variable1; // variable1 is undefined because it has no assigned value
let variable2 = null; // variable2 is intentionally assigned with null to represent the absence of a value

console.log(variable1); // Output: undefined
console.log(variable2); // Output: null
```

```javascript
// Checking the types of the variables
console.log(typeof variable1); // Output: "undefined"
console.log(typeof variable2); // Output: "object" (null is of type "object" in JavaScript)
```

```javascript
// Comparing undefined and null
console.log(variable1 === undefined); // Output: true
console.log(variable2 === null); // Output: true
console.log(variable1 === null); // Output: false
console.log(variable2 === undefined); // Output: false
```

variable1 is declared but not assigned any value, so it is undefined by default. It represents a variable that exists but doesn't have a defined value.

variable2 is assigned with the value null. It intentionally represents the absence of any value. It is different from undefined because null is a value that can be assigned to a variable.

When we log the values of variable1 and variable2 to the console, it displays undefined and null, respectively.

We can also use the typeof operator to check the types of the variables. The typeof variable1 returns "undefined", and typeof variable2 returns "object" because null is of type "object" in JavaScript.

Lastly, we demonstrate the comparison between undefined and null using the strict equality (===) operator. 
Comparing variable1 === undefined and variable2 === null returns true in both cases, indicating that variable1 is undefined and variable2 is null. However, comparing variable1 === null and variable2 === undefined returns false because variable1 is undefined, not null, and vice versa.
