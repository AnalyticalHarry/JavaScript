```javascript
//Addtion of two number
// It take two argument 
function addTwoNumber(num1, num2){
    if(typeof num1 !== 'number' || typeof num1 !== 'number'){
        return false;
    }
    if(typeof num2 !== 'number' || typeof num2 !== 'number'){
        return false;
    }
    let sum = num1 + num2;
    return sum
}
```

```javascript
addTwoNumber(20, 30)
```

```javascript
//This function will help to sum all the list
function sumList(list) {
    var count = 0;
    for (var i = 0; i < list.length; i++) {
        count = count + list[i];
    }
    return count;
}
```

```javascript
sumList([1,2,3,4,5])
```

```javascript
function repeat(str, numTimes){
    let result = '';
    for (let i =0; i<numTimes; i++){
        result = result + str;
    }
    console.log(result);
}
```

```javascript
repeat('Hello ', 5)
```

```javascript
repeat("HarryWorld! ", 10)
```

```javascript
//Days of the Week
function daysOfTheWeek(day){
    if (day === 1){
        console.log("Monday")
    }
    if (day === 2){
        console.log("Tuesday")
    }
    if (day === 3){
        console.log("Wednesday")
    }
    if (day === 4){
        console.log("Thursday")
    }
    if (day === 5){
        console.log("Friday")
    }
    if (day === 6){
        console.log("Saturday")
    }
    if (day === 7){
        console.log("Sunday")
    }
}
```

```javascript
daysOfTheWeek(1)
daysOfTheWeek(2)
daysOfTheWeek(3)
daysOfTheWeek(4)
daysOfTheWeek(5)
daysOfTheWeek(6)
daysOfTheWeek(7)
```

```javascript
// Bubble sort algorithm 
function bubbleSort(List, ascending=true){
    for(var i=0; i<List.length - 1; i++){
        for (var j = 0; j<List.length -i -1; j++){
            if (ascending){
                if(List[j]>List[j+1]){
                    // Swap elements if current element is greater than the next element
                    [List[j], List[j+1]] = [List[j+1], List[j]];
                }
            } else {
                if(List[j]<List[j+1]){
                    // Swap elements if current element is smaller than the next element
                    [List[j], List[j+1]] = [List[j+1], List[j]];
                }
            }
        }
    }
    // Return the sorted list after the loop completes
    return List;
}
```

```javascript
var list = [64, 34, 25, 12, 22, 11, 90];
console.log("Original List:", list);
```

```javascript
bubbleSort(list);
console.log("Sorted List (ascending):", list);
```

```javascript
list = [64, 34, 25, 12, 22, 11, 90];
bubbleSort(list, false);
console.log("Sorted List (descending):", list);
```
